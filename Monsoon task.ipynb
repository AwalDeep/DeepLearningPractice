# Checking correlated features

```python
# Checking feature dependency on other features by plotting heatmap of all features

names = X_train.columns
correlations = X_train.corr()

# plot correlation matrix
plt.rcParams['figure.figsize'] = (20,15)
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1, cmap="viridis")
fig.colorbar(cax)
ticks = np.arange(0,30,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names, rotation=90)
ax.set_yticklabels(names)
plt.show()
```

# There are some highly correlated features one of which we can opt to remove:
# N3 and N4, N2 and N5, N19 and N6, N33 and N34, N8 and N16

```python
# dropping the features based on correlation
X_train.drop('N3',axis=1,inplace=True)
X_train.drop('N5',axis=1,inplace=True)
X_train.drop('N16',axis=1,inplace=True)
X_train.drop('N19',axis=1,inplace=True)
X_train.drop('N33',axis=1,inplace=True)
```

# One Hot Encoding

```python
# Categorical
# c1 to c8
# ['N10','N15','N16','N22','N12','N14']
```

```python
categorical=['C1','C2','C3','C4','C5','C6','C7','C8','N10','N15','N22','N12','N14']
for var in categorical:
    cat_list ='var'+'_'+var
    cat_list = pd.get_dummies(X_train[var], prefix=var)
    data1=X_train.join(cat_list)
    X_train=data1

data_vars=X_train.columns.values.tolist()
to_keep=[i for i in data_vars if i not in categorical]    
```

```python
X_train=X_train[to_keep]
```
